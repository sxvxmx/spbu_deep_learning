Реализуйте медианный фильтр. Для текущего пикселя, пиксели, которые «попадают» в матрицу, сортируются, и выбирается медианное значение из отсортированного массива. Это значение и является выходным для текущего пикселя. Используйте только чистый torch. Покажите результат на примере для размера ядра 3, 5, 10

```python
import torch
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
IMAGE_PATH = "../../data/images"
```

```python
img = Image.open(f"{IMAGE_PATH}/kot_mjaukaet.jpg")
```

```python
img_np = np.array(img)
plt.imshow(img_np)
plt.show()
```

```python
img_np.shape
```

```python
def median_ker(img: np.array, ker_shape: tuple):
    out_img = np.zeros_like(img) 
    for channel in range(img.shape[2]):
        imgp = img[:, :, channel]
        for i in range(imgp.shape[0] - ker_shape[0] + 1): 
            for o in range(imgp.shape[1] - ker_shape[1] + 1):
                out_img[i, o, channel] = np.median(imgp[i:i+ker_shape[0], o:o+ker_shape[1]])
    return out_img
```

```python
keri = median_ker(img_np, (10,10))
```
